```python
ymean=[]
def linear(b0,b1,b2,b3,b4,b5,b6,b7,b8,b9,b10):
  sum=0;
  ybar=[]
  ypp=[]
  for i in range(0,len(x1)):
    ybar.append(b0+b1*x1[i]+b2*x2[i]+b3*x3[i]+b4*x4[i]+b5*x5[i]+b6*x6[i]+b7*x7[i]+b8*x8[i]+b9*x9[i]+b10*x10[i])
  for i in range(0,len(x1)):
    ypp.append((y[i]-ybar[i])**2)
  for i in range(0,len(x1)):
    sum=sum+ypp[i]
  ymean.append(sum/len(x1))
for i in range(0,50):
  linear(i,i+25,i+50,i,i,i,i,i,i,i,i)
```

```python
if(ymean[0]<ymean[1]):
  result=ymean[0]
else:
  result=ymean[1]
for i in range(0,len(ymean)):
  if(result>ymean[i]):
    result=ymean[i]
    
print("the less mean square error is",result)
```
